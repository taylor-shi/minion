```python
def ConvertToFasta(directory):            #path to raw fastq files
    for file in os.listdir(directory+'/fastq'):
        if file.endswith('.fastq'): new_file=file[:-6]+'.fasta'
            SeqIO.convert(directory+'/fastq/'+file,"fastq",directory+'/fasta/'+new_file,"fasta")
```

```python
directory='2019-8-6_cdna/fasta' #Define this directory every time demultiplex() is run

def demultiplex(files):
    for file in files:
        fileNumber = file[:-6]
        
        file01 = open(directory+'/'+str(fileNumber)+'liver.fasta','w')
        file02 = open(directory+'/'+str(fileNumber)+'a431utc.fasta','w')
        file03 = open(directory+'/'+str(fileNumber)+'a431.a1.fasta','w')
        file04 = open(directory+'/'+str(fileNumber)+'a431.b1.fasta','w')
        file05 = open(directory+'/'+str(fileNumber)+'a431.b2.fasta','w')
        file06 = open(directory+'/'+str(fileNumber)+'a431.b3.fasta','w')
        nobarcode = open(directory+'/'+str(fileNumber)+'nobarcode.fasta','w')
       
        with open(directory+'/'+file) as file:
            for read in SeqIO.parse(file,'fasta'):   
                #01
                m = regex.search('(AAGAAAGTTGTCGGTGTCTTTGTG){e<=6}', str(read.seq)[0:101]) #allows 6 barcode mismatches
                if m: 
                    SeqIO.write(read, file01, 'fasta') 
                    continue
                #02
                m = regex.search('(TCGATTCCGTTTGTAGTCGTCTGT){e<=6}', str(read.seq)[0:101])
                if m: 
                    SeqIO.write(read,file02, 'fasta')
                    continue
                #03
                m = regex.search('(GAGTCTTGTGTCCCAGTTACCAGG){e<=6}',str(read.seq)[0:101])
                if m: 
                    SeqIO.write(read,file03, 'fasta')
                    continue
                #04
                m = regex.search('(TTCGGATTCTATCGTGTTTCCCTA){e<=6}',str(read.seq)[0:101])
                if m: 
                    SeqIO.write(read,file04, 'fasta')
                    continue
                #05
                m = regex.search('(CTTGTCCAGGGTTTGTGTAACCTT){e<=6}',str(read.seq)[0:101])
                if m: 
                    SeqIO.write(read,file05, 'fasta')
                    continue
                #06
                m = regex.search('(TTCTCGCAAAGGCAGAAAGTAGTC){e<=6}',str(read.seq)[0:101])
                if m: 
                    SeqIO.write(read,file06, 'fasta')
                    continue
                else: SeqIO.write(read,nobarcode, 'fasta')
                    
def multiRunWrapper(args): # intermediate for parallel processing
    # print 'step 1'
    return demultiplex(args)
def RunMany(inputList): # set up parallel processing, 
    #import psutil
    #cpuNumber = psutil.cpu_count()
    #print 'number of processors:', cpuNumber
    pool = Pool(processes=32)
    imageScores = pool.map(multiRunWrapper, inputList)
    pool.close()
    pool.join()
    return imageScores
    print 'done'
    
def MergeFasta(directory,
              samplelist):    #List of sample (ie. fasta file) names from a sequence run
    print 'Breakdown of yield:'
    for sample in samplelist:
        !cat {directory}/*{sample}.fasta > {directory}/{sample}.fasta
        !grep ^'>' -c {directory}/{sample}.fasta
```
